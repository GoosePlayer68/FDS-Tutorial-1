```python
import numpy as np
arr=np.array([1,2,3])
brr=np.array([4,5,6])
crr=np.hstack((arr,brr))
print(crr)
```

```python
import numpy as np
arr=np.array([1,2,3])
brr=np.array([4,5,6])
crr=np.stack((arr,brr),axis=1)
print(crr)
```

```python
import numpy as np
arr=np.array([1,2,3,4,5,6])
brr=np.array_split(arr,3)
print(brr)
```
